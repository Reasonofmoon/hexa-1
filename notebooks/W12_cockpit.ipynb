## Step 0: 환경 설정

```python
import subprocess
subprocess.run(['apt-get','-qq','-y','install','fonts-nanum'],capture_output=True)
!pip install -q pandas matplotlib
import pandas as pd, numpy as np, matplotlib.pyplot as plt, matplotlib.font_manager as fm
nanum=next((f.fname for f in fm.fontManager.ttflist if 'Nanum' in f.name),None)
if nanum: plt.rcParams['font.family']=fm.FontProperties(fname=nanum).get_name()
plt.rcParams['axes.unicode_minus']=False
print('✅ 준비 완료')
```

## Step 1: 12주 성과 데이터 입력 (✏️ 실제 결과 입력)

```python
RESULTS = {
    'company': '✏️ [교육 기업명]',
    'period': '2026.01 ~ 2026.03',
    'before': {'defect_rate': 4.8, 'oee': 76.2, 'manual_hours_week': 12},
    'after':  {'defect_rate': 2.1, 'oee': 87.5, 'manual_hours_week': 2},
    'systems_built': [
        'AI 불량 분류기 (정확도 94%)',
        '납품/HR 문서 자동화 (월 30건)',
        'KPI 대시보드 + Slack 실시간 알림',
        '예지보전 이상감지 (3σ)',
        '협력사 커뮤니케이션 자동화',
    ],
}
print('✅ 성과 데이터 로드')
```

## Step 2: 성과 시각화 (경영진 보고 차트)

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# 불량률 개선
axes[0].bar(['도입 전','도입 후'],
            [RESULTS['before']['defect_rate'],RESULTS['after']['defect_rate']],
            color=['#e74c3c','#2ecc71'],width=0.5)
axes[0].set_title('불량률 개선 (%)')
improve_d=round(RESULTS['before']['defect_rate']-RESULTS['after']['defect_rate'],1)
axes[0].text(0.5,max(RESULTS['before']['defect_rate'],RESULTS['after']['defect_rate'])*0.5,
             f'▼{improve_d}%p',ha='center',color='white',fontweight='bold',fontsize=14)

# OEE 향상
axes[1].bar(['도입 전','도입 후'],
            [RESULTS['before']['oee'],RESULTS['after']['oee']],
            color=['#e74c3c','#2ecc71'],width=0.5)
axes[1].set_title('OEE 향상 (%)')
improve_o=round(RESULTS['after']['oee']-RESULTS['before']['oee'],1)
axes[1].text(0.5,RESULTS['before']['oee']*0.5,
             f'▲{improve_o}%p',ha='center',color='white',fontweight='bold',fontsize=14)

# 업무 시간 절감
axes[2].bar(['도입 전','도입 후'],
            [RESULTS['before']['manual_hours_week'],RESULTS['after']['manual_hours_week']],
            color=['#e74c3c','#2ecc71'],width=0.5)
axes[2].set_title('주간 수작업 시간 (h)')
save_h=RESULTS['before']['manual_hours_week']-RESULTS['after']['manual_hours_week']
axes[2].text(0.5,RESULTS['before']['manual_hours_week']*0.6,
             f'▼{save_h}h/주',ha='center',color='white',fontweight='bold',fontsize=14)

plt.suptitle(f'{RESULTS["company"]} AX 도입 12주 성과 보고',fontsize=14,fontweight='bold')
plt.tight_layout()
plt.savefig('ax_results.png',dpi=150,bbox_inches='tight')
plt.show()
print('💾 성과 차트 저장 완료')
```

## Step 3: 경영진 보고서 자동 생성

```python
import datetime
systems=chr(10).join([f'  - {s}' for s in RESULTS['systems_built']])
report=f"""# {RESULTS['company']} AI 전환 (AX) 성과 보고서
기간: {RESULTS['period']} / 작성일: {datetime.date.today()}

## 핵심 성과 요약
| 지표 | 도입 전 | 도입 후 | 개선율 |
|---|---|---|---|
| 불량률 | {RESULTS['before']['defect_rate']}% | {RESULTS['after']['defect_rate']}% | ▼{round((RESULTS['before']['defect_rate']-RESULTS['after']['defect_rate'])/RESULTS['before']['defect_rate']*100,1)}% |
| OEE | {RESULTS['before']['oee']}% | {RESULTS['after']['oee']}% | ▲{round((RESULTS['after']['oee']-RESULTS['before']['oee'])/RESULTS['before']['oee']*100,1)}% |
| 주간 수작업 | {RESULTS['before']['manual_hours_week']}h | {RESULTS['after']['manual_hours_week']}h | ▼{round((RESULTS['before']['manual_hours_week']-RESULTS['after']['manual_hours_week'])/RESULTS['before']['manual_hours_week']*100,1)}% |

## 구축된 AI 시스템
{systems}

## 결론
12주 AX 마스터클래스를 통해 현장 데이터 기반 AI 운영 체계를 완성했습니다.
"""
with open('ax_final_report.md','w',encoding='utf-8') as f: f.write(report)
print(report)
from google.colab import files
files.download('ax_final_report.md')
files.download('ax_results.png')
print('✅ 최종 보고서 다운로드 완료!')
print('\n🎉 12주 hexa-1 제조업 AX 마스터클래스 완주!')
```
