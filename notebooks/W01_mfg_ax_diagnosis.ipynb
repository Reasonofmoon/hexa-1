## Step 0: 환경 설정

```python
import subprocess
subprocess.run(['apt-get','-qq','-y','install','fonts-nanum'],capture_output=True)
!pip install -q google-generativeai pandas matplotlib
import matplotlib.pyplot as plt, matplotlib.font_manager as fm, pandas as pd
_n=[f for f in fm.findSystemFonts() if 'Nanum' in f]
if _n: fm.fontManager.addfont(_n[0]); plt.rcParams['font.family']=fm.FontProperties(fname=_n[0]).get_name()
plt.rcParams['axes.unicode_minus']=False
try:
    from google.colab import userdata; API_KEY=userdata.get('GEMINI_API_KEY')
except: API_KEY=input('GEMINI_API_KEY: ')
import google.generativeai as genai; genai.configure(api_key=API_KEY)
model=genai.GenerativeModel('gemini-2.0-flash'); print('✅ 준비 완료')
```

## Step 1: 공장 정보 입력 (✏️)

```python
INFO={'name':'✏️ [교육 기업명/공장명]','site_type':'✏️ [자동차|전자|식품|화학|기타]','employees':'✏️ [직원수]'}
print('✅',INFO['name'])
```

## Step 2: AX 진단 (10개 항목, 각 1~5점)

| 점수 | 의미 |
|------|------|
|1|미도입|3|부분적|5|완전자동화|

```python
SCORES={
    '설비 자동화':1,       # ✏️
    '품질 검사 AI':1,      # ✏️
    'OEE 분석':1,         # ✏️
    '예측 정비':1,        # ✏️
    '재고 관리':1,        # ✏️
    '납기 분석':1,        # ✏️
    'HR 자동화':1,        # ✏️
    '공급망 관리':1,      # ✏️
    'KPI 대시보드':1,     # ✏️
    '생산계획 AI':1,      # ✏️
}
total=sum(SCORES.values())
level='🟢 Starter' if total<25 else ('🔵 Growing' if total<40 else '🟣 AI-Ready')
print(f'총점: {total}/50점 — {level}')
```

## Step 3: 시각화

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots(figsize=(10,5))
clr=['#e74c3c' if v<=2 else '#f39c12' if v<=3 else '#2ecc71' for v in SCORES.values()]
ax.barh(list(SCORES.keys()),list(SCORES.values()),color=clr)
ax.axvline(3,color='gray',linestyle='--',alpha=0.6,label='기준(3점)')
ax.set_xlim(0,5); ax.set_title(f'{INFO["name"]} AX 진단 — {level}'); ax.legend()
plt.tight_layout(); plt.savefig('ax_diag.png',dpi=150,bbox_inches='tight'); plt.show()
```

## Step 4: Gemini 맞춤 개선 로드맵

```python
weak=[k for k,v in SCORES.items() if v<=2]
p=f"""제조업 AI 전환 컨설턴트로서 맞춤 전략:
공장: {INFO['name']} ({INFO['site_type']}) / 총점: {total}/50 ({level})
즉시 개선 항목: {', '.join(weak) if weak else '전반적 고도화'}
12주 로드맵 + 즉시 실행 3가지 + 기대 ROI. 마크다운."""
resp=model.generate_content(p)
print(resp.text)
```

## Step 5: 다운로드

```python
with open('ax_report.md','w',encoding='utf-8') as f:
    f.write(f'# {INFO["name"]} AX 진단 리포트\n총점:{total}/50 {level}\n\n')
    f.write(resp.text)
from google.colab import files
files.download('ax_report.md'); files.download('ax_diag.png')
print('✅ 완료!')
```
