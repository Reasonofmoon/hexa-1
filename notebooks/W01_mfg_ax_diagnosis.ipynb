````python
import io
import os
from pathlib import Path
import json
import zipfile
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager as fm

try:
    from google.colab import files
except Exception:
    files = None

try:
    import google.generativeai as genai
except Exception:
    genai = None

GEMINI_MODEL = "gemini-2.0-flash"

for f in fm.findSystemFonts():
    try:
        fm.fontManager.addfont(f)
    except Exception:
        continue


def safe_upload():
    if files is None:
        print("Colab 환경이 아니라 files.upload를 사용할 수 없습니다.")
        return {}
    try:
        return files.upload()
    except Exception as e:
        print(f"files.upload() 취소/실패: {e}")
        print("files.upload() 취소되어 기본 데이터 사용")
        return {}


def safe_download(filename):
    if files is None:
        print(f"Download skipped in non-Colab: {filename}")
        return
    try:
        files.download(filename)
    except Exception as e:
        print(f"files.download 실패: {e}")


def use_gemini(prompt: str, fallback: str):
    if not genai:
        return fallback
    key = os.getenv("GOOGLE_API_KEY", "").strip()
    if not key:
        return fallback
    try:
        genai.configure(api_key=key)
        model = genai.GenerativeModel(GEMINI_MODEL)
        out = model.generate_content(prompt)
        return getattr(out, "text", "").strip() or fallback
    except Exception as e:
        print(f"Gemini 호출 실패: {e}")
        return fallback

df = pd.DataFrame({
    "항목": list(criteria.keys()),
    "점수": list(map(int, criteria.values())),
})


def score_color(score):
    if score <= 2:
        return "#d62728"
    if score == 3:
        return "#ff7f0e"
    return "#2ca02c"

colors = [score_color(s) for s in df["점수"]]

fig, ax = plt.subplots(figsize=(10, 4))
ax.barh(df["항목"], df["점수"], color=colors)
ax.set_xlim(0, 5)
ax.set_title("AX 진단 점수")
ax.set_xlabel("점수 (1~5)")
for i, v in enumerate(df["점수"]):
    ax.text(v + 0.05, i, str(v), va="center")
plt.tight_layout()
plt.show()

score_summary = df["점수"].agg(["mean", "min", "max"]).round(2).to_dict()
print(score_summary)
````

````python
import io
import os
from pathlib import Path
import json
import zipfile
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import font_manager as fm

try:
    from google.colab import files
except Exception:
    files = None

try:
    import google.generativeai as genai
except Exception:
    genai = None

GEMINI_MODEL = "gemini-2.0-flash"

for f in fm.findSystemFonts():
    try:
        fm.fontManager.addfont(f)
    except Exception:
        continue


def safe_upload():
    if files is None:
        print("Colab 환경이 아니라 files.upload를 사용할 수 없습니다.")
        return {}
    try:
        return files.upload()
    except Exception as e:
        print(f"files.upload() 취소/실패: {e}")
        print("files.upload() 취소되어 기본 데이터 사용")
        return {}


def safe_download(filename):
    if files is None:
        print(f"Download skipped in non-Colab: {filename}")
        return
    try:
        files.download(filename)
    except Exception as e:
        print(f"files.download 실패: {e}")


def use_gemini(prompt: str, fallback: str):
    if not genai:
        return fallback
    key = os.getenv("GOOGLE_API_KEY", "").strip()
    if not key:
        return fallback
    try:
        genai.configure(api_key=key)
        model = genai.GenerativeModel(GEMINI_MODEL)
        out = model.generate_content(prompt)
        return getattr(out, "text", "").strip() or fallback
    except Exception as e:
        print(f"Gemini 호출 실패: {e}")
        return fallback

roadmap_prompt = (
    f"공장명:{plant_info['name']}
"
    f"규모:{plant_info['employees']}명, 유형:{plant_info['site_type']}
"
    f"항목별 점수: {criteria}
"
    f"총점:{score_summary['mean']}
"
    "10개 항목을 기준으로 취약 항목 우선순위(Top 3)와 90일 실행 로드맵을 제시해줘. "
    "각 액션은 담당부서, 산정효과, 주의사항까지 포함한 한국어 JSON 배열로 반환."
)

roadmap = use_gemini(roadmap_prompt, "우선 설비 자동화 연계 데이터 수집, 예측정비 PoC, HR 자동화(근태/점검 이력 연계) 순으로 90일 내 단기 개선 권고")
print(roadmap)

Path("outputs").mkdir(exist_ok=True)
Path("outputs/mfg_ax_diagnosis_report.md").write_text(
    f"# {plant_info['name']} AX 자가진단

{roadmap}

점수요약: {score_summary}
",
    encoding="utf-8"
)

fig_path = "outputs/mfg_ax_diagnosis_chart.png"
plt.figure(figsize=(10, 4))
plt.barh(df["항목"], df["점수"], color=colors)
plt.xlim(0, 5)
plt.title("AX 진단 점수")
plt.tight_layout()
plt.savefig(fig_path, dpi=150)
print(f"차트 저장: {fig_path}")
````
